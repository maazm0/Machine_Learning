```python
import requests
wiki_link = 'https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area'
```

```python
link = requests.get(wiki_link).text
```

```python
print(link)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(link,'lxml')
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
soup.title.string
```

```python
#finding a table with its name 
wiki_table = soup.find('table','wikitable sortable')
```

```python
#extracting all the links in it
links = wiki_table.find_all('a')
```

```python
for link in links:
    print(link.get('href'))
```

```python
#let's import wikitable in out notebook
```

```python
data = []
for row in wiki_table.find_all('tr'):
    row_data=[]
    for cell in row.find_all('td'):
        row_data.append(cell.text)
    data.append(row_data)
```

```python
#converting it into a pandas a DataFrame
import pandas as pd
DF = pd.DataFrame(data,columns=['Rank','Country','% Total','Area','Remark']).drop(0,axis=0).drop(['Remark'],axis=1)
```
