```python
from sklearn.datasets import load_iris
import numpy as np
```

```python
data = load_iris()
```

```python
data.feature_names
```

```python
data.target_names
```

```python
#Let's perform K-means clustering this dataset
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(data.data)
y = data.target
```

```python
from sklearn.cluster import KMeans
```

```python
#Will try elbow technique as well as silhouett score to find optimal number of clusters
```

```python
inertia = []
for k in range(1,11):
    algorithm = KMeans(n_clusters=k)
    algorithm.fit(X)
    inertia.append(algorithm.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
x = [x for x in range(1,11)]
plt.plot(x,inertia)
plt.xlabel("Number of Clusters K")
plt.ylabel('WSS')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil_score = []
for k in range(2,11):
    algorithm = KMeans(n_clusters=k)
    algorithm.fit(X)
    labels = algorithm.labels_
    sil_score.append(silhouette_score(X,labels))
```

```python
plt.plot(range(2,11),sil_score)
plt.grid()
plt.xlabel("Number of Clusters K")
plt.ylabel('Silhouette Score')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
labels = kmeans.labels_
```

```python
plt.scatter(X[labels==0,0],X[labels==0,1],marker='o',color='red',s=50)
plt.scatter(X[labels==1,0],X[labels==1,1],marker='o',color='green',s=50)
plt.scatter(X[labels==2,0],X[labels==2,1],marker='o',color='blue',s=50)
plt.show()
```
