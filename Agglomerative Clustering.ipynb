```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
X = data.data
```

```python
#making a dendrogram to find the optimal number of clusters 
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
```

```python
# as we know its 3clusters for Iris and its also visible  
```

```python
model = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
model.fit(X)
labels = model.labels_
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[labels==0,0],X[labels==0,1],marker='o',color='red',s=50)
plt.scatter(X[labels==1,0],X[labels==1,1],marker = 'o',color = 'green',s=50)
plt.scatter(X[labels==2,0],X[labels==2,1],marker = 'o',color = 'blue',s=50)
plt.show()
```

```python
# last to columns
plt.scatter(X[labels==0,2],X[labels==0,3],marker='o',color='red',s=50)
plt.scatter(X[labels==1,2],X[labels==1,3],marker = 'o',color = 'green',s=50)
plt.scatter(X[labels==2,2],X[labels==2,3],marker = 'o',color = 'blue',s=50)
plt.show()
```
