```python
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk import word_tokenize
import re # regular expression class
from nltk.stem import WordNetLemmatizer # lemmatize is to reduce words with inflected or variant forms like present, past, future 
#or gerund

stemmer = WordNetLemmatizer()
```

```python
data = pd.read_csv(r'C:\Users\Mazz Mirza\Mazz\Study\Datasets\fake_or_real_news.csv')
```

```python
data.head()
```

```python
data.drop(['Unnamed: 0','title'],inplace=True,axis=1) 
```

```python
corpus=[]
for i in range(0,500): #for simplicity let's consider 500 records 
    document=re.sub(r'\s[a-zA-Z]\s'," ",data.text[i]) #removing sinle character
    document = re.sub(r'\^[a-zA-Z]\s+'," ",document) #removing single character in the beginning
    document = re.sub(r'\W'," ",document) # removing special characters
    document = re.sub(r'\s+',' ',document) # removing extra spaces
    document = document.lower()
    document = document.split() # splitting unlike method word_tokenize which will tokenize even special characters
    document = [stemmer.lemmatize(word) for word in document]
    # lemmatize is to reduce words with inflected or variant forms present, past, future or gerund
    document = " ".join(document)
    corpus.append(document)
```

```python
corpus[0]
```

```python
X = corpus
y = data.label[:500]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=123)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
count_vectorizer = CountVectorizer(max_features=1500,min_df=5,max_df=.70,stop_words=stop_words)
```

```python
count_train = count_vectorizer.fit_transform(X_train)
count_test = count_vectorizer.transform(X_test)
```

```python
pd.DataFrame(count_train.toarray(),columns=count_vectorizer.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1500,min_df=5,max_df=0.70,stop_words=stop_words)
tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)
```

```python
pd.DataFrame(tfidf_train.toarray(),columns=tfidf_vectorizer.get_feature_names_out())
```

```python
from sklearn.naive_bayes import MultinomialNB
nbclassifier = MultinomialNB()
```

```python
nbclassifier.fit(count_train,y_train)
```

```python
pred = nbclassifier.predict(count_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
nbclassifier = MultinomialNB()
nbclassifier.fit(tfidf_train,y_train)
```

```python
pred = nbclassifier.predict(tfidf_test)
```

```python
accuracy_score(y_test,pred)
```
